```python
# load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load CSV as Pandas dataframe
filename = "/home/emma/Documents/research/gnn-dna/figures/mlp-vs-gnn.csv"
df1 = pd.read_csv(filename)
df1
```

```python
# extract mlp-256-train and val
mlp256_train = df1["mlp-256 - train_loss"].dropna()
mlp256_val = df1["mlp-256 - val_loss"].dropna()

# extract gnn-256-knn-7 train and val
gnn256_train = df1["gnn-256-knn-7 - train_loss"].dropna()
gnn256_val = df1["gnn-256-knn-7 - val_loss"].dropna()
```

```python
# plot mlp-256 vs gnn-256-knn-7
x = np.linspace(0,100,100)
plt.plot(x, mlp256_train, "red", label="MLP Train")
plt.plot(x, mlp256_val, "salmon", linestyle="--", label="MLP Validation")
plt.plot(x, gnn256_train, "blue", label="GNN Train")
plt.plot(x, gnn256_val, "cornflowerblue", linestyle="--", label="GNN Validation")
plt.yscale("log")
plt.grid("on")
plt.legend()
plt.savefig("mlp-vs-gnn.png", transparent=True, dpi=150)
```

```python
# plot gnn-64-knn-3 vs -5 vs -7

filename = "/home/emma/Documents/research/gnn-dna/figures/knn-exploration.csv"
df2 = pd.read_csv(filename)
df2
```

```python
k3_train = df2["gnn-64-knn-3 - train_loss"].dropna()
k3_val = df2["gnn-64-knn-3 - val_loss"].dropna()

k5_train = df2["gnn-64-knn-5 - train_loss"].dropna()
k5_val = df2["gnn-64-knn-5 - val_loss"].dropna()

k7_train = df2["gnn-64-knn-7 - train_loss"].dropna()
k7_val = df2["gnn-64-knn-7 - val_loss"].dropna()
```

```python
plt.plot(x, k3_train, "red", label="K = 3 Train")
plt.plot(x, k3_val, "salmon", linestyle="--", label="K = 3 Validation")
plt.plot(x, k5_train, "blue", label="K = 5 Train")
plt.plot(x, k5_val, "cornflowerblue", linestyle="--", label="K = 5 Validation")
plt.plot(x, k7_train, "darkorange", label="K = 7 Train")
plt.plot(x, k7_val, "bisque", linestyle="--", label="K = 7 Validation")
plt.yscale("log")
plt.grid("on")
plt.legend()
plt.savefig("k-exploration.png", transparent=True, dpi=150)
```
